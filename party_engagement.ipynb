```python
import json
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
df = pd.read_csv("condensed_10.csv", parse_dates=["created_at"])
```

```python
df = df.assign(engagement=lambda x: (x.num_replies+x.num_quotes+x.num_retweets)/x["user.followers_count"])
```

```python
df.groupby("author_party",as_index=False)[["engagement"]].mean()
```

```python
df[df.is_politician].groupby(pd.Grouper(key="created_at", freq="d")).apply(lambda x: x.loc[x["engagement"] == x["engagement"].max()]).sort_values(by="engagement", ascending=False).reset_index(level=1, drop=True)
```
